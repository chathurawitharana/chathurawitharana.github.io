```
#Exercise 1

#import libraries
import matplotlib.pyplot as plt
import numpy as np

#create data arrays
x=np.arange(-4,4,0.01)
y=x**3 +1/(x**2 + 1)

#plot the graph
plt.plot(x,y)

#format the figure
plt.title('y=x^3 +1/(x^2 + 1)',fontsize=20)  #give the title to the graph with fontsize=20
plt.grid()
plt.xlabel('x axis',fontsize=15)       #give the name for x axis
plt.ylabel('y axis',fontsize=15)
plt.show
```

```
#Exercise 2

#import libraries
from numpy import *
from matplotlib.pyplot import *

#create data arrays
x=arange(-1,1,0.01)
y=x/(1+ log(x))

#plot the graph
plot(x,y,'c')

#format the figure
title('y=x/(1+ ln(x))',fontsize=23)
xlabel('x axis',fontsize=17)
ylabel('y axis',fontsize=17)
grid()
show()
```

```
#Exercise 3

#import libraries
import numpy as np 
import matplotlib.pyplot as plt

#create data arrays
x=np.arange ( -2*np.pi , 2*np.pi, np.pi/24 )
y=np.cos(x)/(1+ np.sin(x))

#plot the graph
plt.plot(x,y,'m')

#format the figure
plt.title('y=cos(x)/(1+sin(x))',{'fontsize':20})
plt.xlabel('angle(radians)',{'fontsize':15})
plt.ylabel('y axis',{'fontsize':15})
plt.grid()
plt.show()
```
